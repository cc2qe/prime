Check if number is prime (http://stackoverflow.com/a/23266245)

```r
is.prime <- function(n) n == 2L || all(n %% 2L:ceiling(sqrt(n)) != 0)
```

Goldbach function

```r
goldbach <- function(n, col=c('steelblue', 'indianred')) {
    x <- seq(0,n*pi,0.1)
    
    # blank plot
    plot(1, xlim=c(0,n), ylim=c(-1,1), type='n', xlab='', ylab='')
    abline(h=0)
    
    # from the left
    for (i in 2:n) {
    points(x/(pi/i), sin(x), xlim=c(0,n), type='l', col=col[1])
    }

    # from the right
    for (i in 2:n) {
        points(n - x/(pi/i), sin(x), xlim=c(0,n), type='l', col=col[2])
    }
}

```

```r
n.goldbach <- function(n) {
    for (x in 2:(n%/%2)) {
        print(x)
        if (is.prime(x) & is.prime(n-x)) {
            print(paste(x, n-x))
        }
    }
}
```
