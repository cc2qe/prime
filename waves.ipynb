```r
x <- seq(0,100*pi,0.1)
```

```r
mypal <- c('steelblue', 'indianred')
```

```r
max <- 40

plot(x/(pi/2), sin(x), xlim=c(0,max), type='l', col=mypal[1])
points(x/(pi/3), sin(x), xlim=c(0,max), type='l', col=mypal[1])
points(x/(pi/4), sin(x), xlim=c(0,max), type='l', col=mypal[1])
points(x/(pi/5), sin(x), xlim=c(0,max), type='l', col=mypal[1])
points(x/(pi/6), sin(x), xlim=c(0,max), type='l', col=mypal[1])
points(x/(pi/7), sin(x), xlim=c(0,max), type='l', col=mypal[1])

abline(h=0)
for (i in 0:max) {
    abline(v=i)
}
```
