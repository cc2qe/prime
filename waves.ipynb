```r
x <- seq(0,100*pi,0.1)
```

```r
mypal <- c('steelblue', 'indianred')
```

```r
max <- 30

# plot(x/(pi/2), sin(x), xlim=c(0,max), type='l', col=mypal[1])
# points(x/(pi/1), sin(x), xlim=c(0,max), type='l', col=mypal[1])
plot(1, xlim=c(0,max), ylim=c(-1,1), type='n', xlab='', ylab='')

for (i in 2:max) {
    points(x/(pi/i), sin(x), xlim=c(0,max), type='l', col=mypal[1])
}

abline(h=0)
```
