```r
x <- seq(0,100*pi,0.1)
```

```r
mypal <- c('steelblue', 'indianred')
```

```r
plot(x/(pi/2), sin(x), xlim=c(0,10), type='l', col=mypal[1])
points(x/(pi/3), sin(x), xlim=c(0,10), type='l', col=mypal[1])
points(x/(pi/5), sin(x), xlim=c(0,10), type='l', col=mypal[1])
points(x/(pi/7), sin(x), xlim=c(0,10), type='l', col=mypal[1])

abline(h=0)
for (i in 1:10) {
    abline(v=2)

}
```
